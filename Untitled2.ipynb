```python
#working with rating columns
def clean_rating(rating):
    if rating == "":
        return None
    else:
        return rating

# Apply the function to the 'rating' column
data['rating'] = data['rating'].apply(lambda x: clean_rating(x))
```

```python
#workign on colors
def colors_conversion(df):
    new_rows = []
    for index, row in df.iterrows():
        colors = row['colors']
        if len(colors) > 1:
            for color in colors:
                new_row = row.copy()
                new_row['colors'] = color
                new_rows.append(new_row)
        else:
            new_rows.append(row)
    return pd.DataFrame(new_rows) 
```

```python
# Assuming df is your DataFrame
def size_conversion(df):
    new_rows = []
    for index, row in df.iterrows():
        sizes = row['sizeDetails']
        # Check if sizes is a list
        if isinstance(sizes, list):
            for size in sizes:
                if size != 'Select a size':
                    new_row = row.copy()
                    new_row['sizeDetails'] = size
                    new_rows.append(new_row)
        else:
            if sizes != 'Select a size':
                new_row = row.copy()
                new_row['sizeDetails'] = sizes
                new_rows.append(new_row)
    return pd.DataFrame(new_rows)
```
