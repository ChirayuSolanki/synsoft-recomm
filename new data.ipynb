```python
women_shoe_types = [
    "athletic",
    "booties",
    "boots",
    "clogs",
    "comfort",
    "dress shoes",
    "espadrilles",
    "flats",
    "heels",
    "heel",
    "mules",
    "slides",
    "oxfords",
    "loafers",
    "loafer",
    "platforms",
    "sandals",
    "slippers",
    "sneakers",
    "vacation & resort",
    "pumps"
]

def extract_word(input_string):
    words = input_string.lower().split()
    for word in words:
        if word.lower() in women_shoe_types:
            if(word == 'pumps' or word == 'heels'):
                return 'heel'
            else:
                return word
    return None


# Apply the extract_word function to each element of the 'name' column
data['category'] = data['name'].apply(extract_word)
```

```python
#workign on colors
def colors_conversion(df):
    new_rows = []
    for index, row in df.iterrows():
        colors = row['colors']
        if len(colors) > 1:
            for color in colors:
                new_row = row.copy()
                new_row['colors'] = color
                new_rows.append(new_row)
        else:
            new_rows.append(row)
    return pd.DataFrame(new_rows)
```

```python
# prompt: write me a code to convert '$79.00' in float 79.00 for column exactPrice



data_1['exactPrice'] = data_1['exactPrice'].apply(lambda x: float(x[1:]))
```

```python
def recommend(data_1,size,price):
  rec = data_1[(data_1['sizeDetails'] == size) & (data_1['exactPrice'] == price)]
  if rec.shape[0]<10:
    return rec
  else:
    heel_rows = rec[rec['category'] == 'heel']
    sandals_flats_rows = rec[rec['category'].isin(['sandals', 'flats'])]
    #sort by review then rating
    #price -> price range
    #number of reccommendation
    #({dict of fraction})
    # Select 12 random rows from heel_rows
    heel_sample = heel_rows.sample(n=12)

    # Select 8 random rows from sandals_flats_rows
    sandals_flats_sample = sandals_flats_rows.sample(n=8)

    # Combine the two samples
    sample_df = pd.concat([heel_sample, sandals_flats_sample])

    # Shuffle the rows of the combined sample
    sample_df = sample_df.sample(frac=1).reset_index(drop=True)

    # Display the first 20 rows
    return sample_df.head(20)

```

```python
a = recommend(data_1,'6W',79.00)
```
